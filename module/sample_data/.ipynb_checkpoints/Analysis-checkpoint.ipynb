```python
import openpyxl
from openpyxl import load_workbook
from openpyxl_image_loader import SheetImageLoader
```

```python
wb = openpyxl.load_workbook("sample.xlsx")
```

```python
ws = wb[wb.sheetnames[0]]
image_loader = SheetImageLoader(ws)
```

```python
ws["B3"].value
```

```python
img_locs = []
for image_loc in image_loader._images:
    img_locs.append(image_loc)
```

```python
# clean the image_locs and sort
img_locs_filtered = [item for item in img_locs if item.startswith('A')]
img_locs_ft = sorted(img_locs_filtered, key=lambda x: int(x[1:]))
```

```python
imgs = []
for cell in img_locs_ft:
    imgs.append(image_loader.get(cell))
```

```python
import io
```

```python
stream = io.BytesIO()
imgs[0].save(stream, format="JPEG")
img_byte_arr = stream.getvalue()
```

```python
from PIL import Image
stream = io.BytesIO(img_byte_arr)
img = Image.open(stream)
img
```
